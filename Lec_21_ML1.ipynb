By trying different combination of x1-x2 scatter plots, we seem to find that some x1-x2 could be very useful to separate them.
For instance, the red color sets tend to have small pepal length and width, which the grey color sets tend to have larger values. On the other hand, the sepal length/width pair seems not a good descriptor to distinguish them. Since we only have four features, let's try to list all the possibilities.

````python
import matplotlib.pyplot as plt

label = ['Sepal Length','Sepal Width','Petal Length','Petal Width']


fig, axarr = plt.subplots(4, 4)
fig.set_size_inches(18, 18)

for i in range(4):
    x1 = X[:,i]
    for j in range(4):
        x2 = X[:,j]
        if i == j:
            axarr[i,j].text(0.2, 0.5, label[i],fontsize=20)
        else:
            axarr[i,j].scatter(x1, x2, c=Y, cmap=plt.cm.Set1,edgecolor='k')
            
plt.show()
````

From the above figure, we could already find that petal width/length are more sensitive charteristics than sepal width/length. Therefore, we could naively set a classification model as follows,
- 1, if petal length is between 0.0-2.4 and petal width is between 0.0-0.8, it is red (setosa)
- 2, if petal length is between 2.4-5.0 and petal width is between 0.8-1.6, it is orange (versicolor)
- 3, otherwise, it is gray (virginica)


````python
def Iris_classifier(x):

    # a simple calssifier from our eyes
    label = None

    if x[3]>0 and x[3]<0.8 and x[2]>0 and x[2]<2:
        label = 0 #'setosa'
    elif x[3]>0.8 and x[3]<1.6 and x[2]>3 and x[2]<5:
        label = 1 #'versicolor'
    else:
        label = 2 #'virginica'

    return label
````

````python
# Let's try to calculate the success rate of this classfier
count = 0
for xx, yy in zip(X, Y):
    if Iris_classifier(xx)== yy:
        count += 1
        
print('The overall success rate is: ', count/len(X)*100 ,'%')
````

### 23.4 Test and Validation

When we work on the machine learning model, it is usually not a good idea to use all the dataset to build up the model.

Insteads, it is much better to evaluate predictive models by partitioning the original sample into a _training set_ to train the model, and a _test set_ to evaluate it. This is called _Cross-validation_.

The partitioning can be very straightfoward, say, we randomly pick 2/3 of the data for training and the rest 1/3 for testing. However, it is better use the built_in function from scikit-learn as follows. 

````python
from sklearn.cross_validation import train_test_split
Xd_train, Xd_test, Y_train, Y_test = train_test_split(X, Y)

def success_ratio(X_test, Y_test):
    count = 0
    for xx, yy in zip(X_test, Y_test):
        if Iris_classifier(xx) == yy:
            count += 1
    return float(count/len(X_test))

print(success_ratio(Xd_test, Y_test))
````

### 23.5 One Rule Algorithm

In the above section, we have developed a simple classifier based on the graph analysis, which could almost reach 100 % of success rate. However, this is not machine learning! What we really want is to let computer find the rule automatically. 
So we need a more automtated algorithm.

#### 23.5.1 Discetization

Unlike human, computers like discrete number!
It is usally useful to transform the data from continuous numbers to the discrete numbers. The simplest way is to 
take the average value, set all the values which are lower than the mean as 0, and all the values higher than mean as 1. This could be simply done by one line with the help of numpy.

````python
import numpy as np
X_d = np.array( X>X.mean(), dtype='int')
````

````python

from collections import OrderedDict
import pandas as pd

Iris_data_d = OrderedDict({'Sepal Length ': X_d[:,0],
             'Sepal Width': X_d[:,1],
             'Petal Length': X_d[:,2],
             'Petal Width': X_d[:,3],
             'Label': Y})

df = pd.DataFrame(Iris_data_d)
df
````

After we have the table, we can apply "One Rule" algorithm based on the frequency, which a simple, yet accurate, classification algorithm based on one rule. 

The algorithm is the following. 
- 1, For each predictor,
     - 1.1 For each value of that predictor, make a rule as follows;
           - 1.1.1 Count how often each value of target (class) appears
           - 1.1.2 Find the most frequent class
           - 1.1.3 Make the rule assign that class to this value of the predictor
     - 1.2 Calculate the total error of the rules of each predictor
- 2, Choose the predictor with the smallest total error.
