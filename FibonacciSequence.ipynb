```python
#The Fibonacci sequence is 
#1, 1, 2, 3, 5, 8, 13, 21, 34,...
#We can write it using the recurrence relation
#f(n+1) = f(n) + f(n-1) with the initial conditions f(1) = f(2) = 1.
#This program will output the first 100 Fibonacci numbers.
```

```python
i = 3 #This variable represents the position of the Fibonacci number.
f1 = f2 = 1 #These are the initial conditions for the Fibonacci sequence.
fn = 0 #This variable represents an arbitrary Fibonacci number.

print("1 -> ", f1) #The first two Fibonacci numbers are printed here.
print("2 -> ", f2)

#This while loop calculates each Fibonacci number and prints them.
while i < 101:
    fn = f1 + f2
    f2 = f1
    f1 = fn
    print(i,"-> ", fn)
    i += 1
    
    
```
